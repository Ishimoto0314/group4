```python
#librosaをimport
import librosa

#サンプルOGGファイルのパスを取得
file_name = librosa.util.example_audio_file()

print(file_name)
```

```python
#先ほどのパスを使用してロード
# y：波形
# sr：サンプリングレート

y, sr = librosa.load(file_name)
print(type(y)) 
print(y.shape) 
print(type(sr), sr) 
```

```python
#フレームの取得
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
print(beat_frames)
```

```python
#ビートのタイミングを時刻表示
eat_times = librosa.frames_to_time(beat_frames, sr=sr)

print(beat_times)
```
